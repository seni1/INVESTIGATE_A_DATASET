> **Approach**: First I create the decade columns using mean values.


```python
#Calculating the mean and creating columns for 10year periods
def decades(dx):
    
    dx['80s'] = dx.iloc[:, 6:17].mean(axis=1)
    dx['90s'] = dx.iloc[:, 16:27].mean(axis=1)
    dx['00s'] = dx.iloc[:, 26:37].mean(axis=1)
    dx['10s'] = dx.iloc[:, 36:47].mean(axis=1)
    
decades(w_1524)
decades(w_2534)
decades(w_3544)
decades(w_4554)
decades(w_5564)
decades(w_65)
```
